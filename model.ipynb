```python
#importing libraries
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
```

```python
#loading the data and scaling the data
(train_im, train_la), (test_im, test_la) = datasets.cifar10.load_data()
train_im, test_im = train_im/255.0, test_im/255.0
```

```python
#creating a convolutional neural network
model = models.Sequential()
model.add(layers.Conv2D(32,(3,3), activation='relu', input_shape=(32,32,3))) #rgb image
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64,(3,3), activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64,(3,3), activation='relu'))
model.summary()
```

```python
#now flattening the data and adding a fully connected neural network
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
model.summary()
```

```python
#adding the optimisers, metrics
model.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(),metrics=['accuracy'])
```

```python
predict = model.fit(train_im, train_la, epochs=10, validation_data=(test_im, test_la))
```

```python
#testing accuracy
test_loss, test_acc = model.evaluate(test_im, test_la, verbose=2)
print(test_acc)
```
